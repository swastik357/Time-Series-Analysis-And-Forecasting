```python
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0 = np.random.randn(1000)
```

```python
plt.plot(x0);
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
plot_pacf(x0, ax=ax);
```

```python
x1 = [0]
for i in range(1000): x1.append(0.5 * x1[-1] + 0.1 * np.random.randn())
x1 = np.array(x1)
```

```python
plt.plot(x1);
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
plot_pacf(x1, ax=ax);
```

```python
# iid noise
fig, ax = plt.subplots(figsize=(10, 5))
plot_acf(np.random.randn(1000), ax=ax);
```
