```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import mean_absolute_percentage_error, r2_score
```

```python
df0 = pd.read_csv('sp500sub.csv', index_col='Date', parse_dates=True)
```

```python
df0.head()
```

```python
df = df0[df0['Name'] == 'IBM'][['Close']].copy()
```

```python
df['LogClose'] = np.log(df['Close'])
```

```python
df['DiffLogClose'] = df['LogClose'].diff()
```

```python
Ntest = 12
train = df.iloc[:-Ntest]
test = df.iloc[-Ntest:]
```

```python
# Make supervised dataset
# let's see if we can use T past values to predict the next value

series = df['DiffLogClose'].to_numpy()[1:] # since 1st row of df['DiffLogClose'] is NAN

T = 21
X = []
Y = []
for t in range(len(series) - T):
  x = series[t:t+T]
  X.append(x)
  y = series[t+T]
  Y.append(y)

X = np.array(X).reshape(-1, T)
Y = np.array(Y)
N = len(X)
print("X.shape", X.shape, "Y.shape", Y.shape)
```

```python
Xtrain, Ytrain = X[:-Ntest], Y[:-Ntest]
Xtest, Ytest = X[-Ntest:], Y[-Ntest:]
```

```python
lr = LinearRegression()
lr.fit(Xtrain, Ytrain)
lr.score(Xtrain, Ytrain)
```

```python
lr.score(Xtest, Ytest)
```

```python
# Boolean index
train_idx = df.index <= train.index[-1]
test_idx = ~train_idx

train_idx[:T+1] = False # first T+1 values are not predictable
```

```python
# Needed to compute un-differenced predictions
df['ShiftLogClose'] = df['LogClose'].shift(1)
prev = df['ShiftLogClose']
```

```python
# Last-known train value
last_train = train.iloc[-1]['LogClose']
```

```python
# 1-step forecast
df.loc[train_idx, 'LR_1step_train'] = prev[train_idx] + lr.predict(Xtrain) # since pred(LogClose(t)) = true(LogClose(t-1)) + pred('DiffLogClose'(t))
df.loc[test_idx, 'LR_1step_test'] = prev[test_idx] + lr.predict(Xtest)
```

```python
# plot 1-step forecast
cols = ['LogClose', 'LR_1step_train', 'LR_1step_test']
df.iloc[-100:][cols].plot(figsize=(15, 5));
```

```python
# multi-step forecast
multistep_predictions = []

# first test input
last_x = Xtest[0]

while len(multistep_predictions) < Ntest:
  p = lr.predict(last_x.reshape(1, -1))[0]

  # update the predictions list
  multistep_predictions.append(p)

  # make the new input
  last_x = np.roll(last_x, -1) # we discard oldest data and append NEWEST PREDICTION
  last_x[-1] = p
```

```python
# save multi-step forecast to dataframe
df.loc[test_idx, 'LR_multistep'] = last_train + np.cumsum(multistep_predictions) # since pred(LogClose(t)) = true_last_train(LogClose) + cumulative sum of (pred('DiffLogClose'(t)))
```

```python
# plot 1-step and multi-step forecast
cols = ['LogClose', 'LR_multistep', 'LR_1step_test']
df.iloc[-100:][cols].plot(figsize=(15, 5));
```

```python
# make multi-output supervised dataset
Tx = T
Ty = Ntest
X = []
Y = []
for t in range(len(series) - Tx - Ty + 1): # using series(t),series(t+1),,,series(t+T-1) to predict series(t+T),series(t+T+1),..,series(t+T+Ntest-1) in ONE-SHOT
  x = series[t:t+Tx]
  X.append(x)
  y = series[t+Tx:t+Tx+Ty]
  Y.append(y)

X = np.array(X).reshape(-1, Tx)
Y = np.array(Y).reshape(-1, Ty)
N = len(X)
print("X.shape", X.shape, "Y.shape", Y.shape)
```

```python
Xtrain_m, Ytrain_m = X[:-1], Y[:-1]
Xtest_m, Ytest_m = X[-1:], Y[-1:] # the test X has only ONE ROW because it predicts series(len(series)-Ntest),series(len(series)-Ntest+1),..,series(len(series)-1) in ONE-SHOT
```

```python
lr = LinearRegression()
lr.fit(Xtrain_m, Ytrain_m)
lr.score(Xtrain_m, Ytrain_m)
```

```python
r2_score(Ytest_m.flatten(), lr.predict(Xtest_m).flatten())
```

```python
# save multi-output forecast to dataframe
df.loc[test_idx, 'LR_multioutput'] = last_train + \
  np.cumsum(lr.predict(Xtest_m).flatten())
```

```python
# plot all forecasts
cols = ['LogClose', 'LR_multistep', 'LR_1step_test', 'LR_multioutput']
df.iloc[-100:][cols].plot(figsize=(15, 5));
```

```python
# MAPE
test_log_pass = df.iloc[-Ntest:]['LogClose']
mape1 = mean_absolute_percentage_error(
    test_log_pass, df.loc[test_idx, 'LR_multistep'])
print("multi-step MAPE:", mape1)
mape2 = mean_absolute_percentage_error(
    test_log_pass, df.loc[test_idx, 'LR_multioutput'])
print("multi-output MAPE:", mape2)
```

```python
def one_step_and_multistep_forecast(model, name):
  model.fit(Xtrain, Ytrain)
  print("One-step forecast:", name)
  # print("Train R^2:", model.score(Xtrain, Ytrain))
  # print("Test R^2 (1-step):", model.score(Xtest, Ytest))

  # store 1-step forecast
  df.loc[train_idx, f'{name}_1step_train'] = \
    prev[train_idx] + model.predict(Xtrain)
  df.loc[test_idx, f'{name}_1step_test'] = \
    prev[test_idx] + model.predict(Xtest)

  # generate multi-step forecast
  multistep_predictions = []

  # first test input
  last_x = Xtest[0]

  while len(multistep_predictions) < Ntest:
    p = model.predict(last_x.reshape(1, -1))[0]

    # update the predictions list
    multistep_predictions.append(p)

    # make the new input
    last_x = np.roll(last_x, -1) # we discard oldest data and append NEWEST PREDICTION
    last_x[-1] = p

  # store multi-step forecast
  df.loc[test_idx, f'{name}_multistep_test'] = \
    last_train + np.cumsum(multistep_predictions)

  # MAPE of multi-step forecast
  mape = mean_absolute_percentage_error(
      test_log_pass, df.loc[test_idx, f'{name}_multistep_test'])
  print("Test MAPE (multi-step):", mape)

  # plot 1-step and multi-step forecast
  cols = [
    'LogClose',
    f'{name}_1step_train',
    f'{name}_1step_test',
    f'{name}_multistep_test'
  ]
  df.iloc[-100:][cols].plot(figsize=(15, 5));
```

```python
one_step_and_multistep_forecast(SVR(), "SVR")
```

```python
one_step_and_multistep_forecast(RandomForestRegressor(), "RF")
```

```python
def multi_output_forecast(model, name):
  model.fit(Xtrain_m, Ytrain_m)

  # save multi-output forecast to dataframe
  df.loc[test_idx, f'{name}_multioutput'] = \
    last_train + np.cumsum(model.predict(Xtest_m).flatten())

  mape = mean_absolute_percentage_error(
      test_log_pass, df.loc[test_idx, f'{name}_multioutput'])
  print("Test MAPE (multi-output):", mape)

  # plot all forecasts
  # assume that previous functions have been run
  cols = [
    'LogClose',
    f'{name}_1step_train',
    f'{name}_1step_test',
    f'{name}_multistep_test',
    f'{name}_multioutput',
  ]
  df.iloc[-100:][cols].plot(figsize=(15, 5));
```

```python
'''
multi_output_forecast(SVR(), "SVR") # cannot handle multi-output!
Therefore below we create a SVR wrapper. The SVR wrapper creates ONE SEPARATE SUPPORT VECTOR MACHINE for EACH COLUMN OF Ytrain_m/Ytest_m
So, a total of Ntest SUPPORT VECTOR MACHINES are created.
'''
```

```python
class SVRWrapper:
  def __init__(self, h, **args):
    self.h = h
    self.models = [SVR(**args) for _ in range(h)]

  def fit(self, X, Y):
    for k in range(self.h):
      self.models[k].fit(X, Y[:, k])

  def predict(self, X):
    P = [m.predict(X) for m in self.models]
    return np.hstack(P)
```

```python
multi_output_forecast(SVRWrapper(Ntest), "SVR")
```

```python
multi_output_forecast(RandomForestRegressor(), "RF")
```
